### engineer coordinates

```python
from geopy.geocoders import Nominatim
def get_city_coordinates(city):
    # Create a geocoder instance
    geolocator = Nominatim(user_agent="my_app")
    # Geocode the city to get the location information
    location = geolocator.geocode(city, language="he", timeout=10)
    # Extract the latitude and longitude from the location data
    if location:
        latitude = location.latitude
        longitude = location.longitude
        return longitude, latitude
    return None, None
```

```python
get_city_coordinates('תל אביב יפו')
```

```python
processed_tama_data['longitude'] = processed_tama_data['hebrew_city_name'].apply(lambda city: get_city_coordinates(city)[0])
processed_tama_data['latitude'] = processed_tama_data['hebrew_city_name'].apply(lambda city: get_city_coordinates(city)[1])
processed_tama_data
```

```python
cols = ['hebrew_city_name', 'city_code', 'year', 'total_open_requests', 'total_requests', 'total_approved_requests', 'longitude', 'latitude']
processed_tama_data = processed_tama_data[cols]
processed_tama_data.sort_values(by=cols, ignore_index=True, inplace=True)
processed_tama_data.to_csv('engineered_tama_data.csv', index=False)
processed_tama_data
```
