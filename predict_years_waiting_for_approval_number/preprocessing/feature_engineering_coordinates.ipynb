```python
tama_df = pd.read_csv('tama.csv', encoding='cp1255')
```

```python
tama_df['address'] = tama_df['Rechov'] + ' ' + tama_df['Bayt'].apply(str) + ', ' + tama_df['Yeshuv']
```

```python
from geopy.geocoders import GoogleV3
from functools import partial
from geopy.exc import GeocoderTimedOut

# Replace YOUR_API_KEY with your actual API key
YOUR_API_KEY = 'AIzaSyCoAqXl84XfBj5I3Eyj06CRxiGaKYDZmbI'

# Create a GoogleV3 geocoder object with your API key
geolocator = GoogleV3(api_key=YOUR_API_KEY)

# Define a function to geocode a Hebrew address using the GoogleV3 geocoder with a timeout of 10 seconds
def geocode_hebrew_address(address):
    try:
        location = geolocator.geocode(address, language='he', timeout=10)
        return (location.latitude, location.longitude) if location else None
    except GeocoderTimedOut:
        print(f"Geocoding timed out for address '{address}'.")
        return None
    except Exception as e:
        print(f"Error geocoding address '{address}': {e}")
        return None

# Apply the geocode_hebrew_address function to the 'address' column of your DataFrame
tama_df['coordinates'] = tama_df['address'].apply(geocode_hebrew_address)
tama_df['latitude'] = tama_df['coordinates'].apply(lambda x: x[0] if x else None)
tama_df['longitude'] = tama_df['coordinates'].apply(lambda x: x[1] if x else None)
```
